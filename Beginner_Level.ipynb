````python
# Check if a string is palindrome

'''
Write a function is_palindrome(s) that checks if a given string s is a palindrome (reads the same forwards and backwards).

Example:
Input: "madam" ‚Üí Output: True
Input: "hello" ‚Üí Output: False
'''

def is_palindrome(string):
    if string.lower() == string[::-1].lower():
        return True
    else:
        return False
    
is_palindrome('Ekitike') # True
is_palindrome('Ant')
````

````python
# Count Vowels in a String
'''
Write a function count_vowels(s) that returns the number of vowels (a, e, i, o, u) in a string. 
It should be case-insensitive.

Example:
Input: "Education" ‚Üí Output: 5
'''

vowels = ['a', 'e', 'i', 'o', 'u']
def count_vowels(string):
    count = 0
    string = string.lower()
    for character in string:
        if character in vowels:
            count += 1
        else:
            continue
    return count

count_vowels('Umbrella')
````

````python
# Remove duplicates from a string 

'''
Write a function remove_duplicates(s) that removes duplicate characters from a string while preserving the original order.

Example:
Input: "banana" ‚Üí Output: "ban"
'''

def remove_duplicates(string):
    string = list(set(string.lower()))
    return ''.join(string)

remove_duplicates('banana') 


# Alternate solution

def alt_remove_duplicates(string):
    string_list = []
    for character in string.lower():
        if character in string_list:
            continue
        else:
            string_list.append(character)

    return ''.join(string_list)

alt_remove_duplicates('banana')
````

````python
# Find the most frequent character

'''
Write a function most_frequent_char(s) that returns the character that appears the most in a string (excluding spaces).

Example:
Input: "hello world" ‚Üí Output: "l"
'''


def most_frequent_char(string):

    strings_dictionary = {}
    for character in string:
        if character != ' ':
            if character in strings_dictionary:
                strings_dictionary[character]  += 1
            else:
                strings_dictionary[character] = 1

    for character, count in strings_dictionary.items():
        if count == max(strings_dictionary.values()):
            print(character)

most_frequent_char('jeff bell')
````

````python
# Reverse words in a sentence

'''
Write a function reverse_words(sentence) that takes a sentence and returns the words in reverse order.

Example:
Input: "I love Python" ‚Üí Output: "Python love I"
'''

def reverse_words(sentence):
    words = sentence.split(' ')
    words.reverse() # alternate -> words = words[::-1]
    return ' '.join(words)

reverse_words('I love FCBarcelona and Messi')
````

Always try to catch specific exceptions to make your code clearer and safer.

The else block runs only if no exception was raised.
